```python
import requests
import json
import config
# Make Strava auth API call with your 
# client_code, client_secret and code
response = requests.post(
                    url = 'https://www.strava.com/oauth/token',
                    data = {
                            'client_id': config.client_ID,
                            'client_secret': config.client_secret,
                            'code': code,
                            'grant_type': 'authorization_code'
                            }
                )
#Save json response as a variable
strava_tokens = response.json()
# Save tokens to file
with open('strava_tokens.json', 'w') as outfile:
    json.dump(strava_tokens, outfile)
# Open JSON file and print the file contents 
# to check it's worked properly
with open('strava_tokens.json') as check:
  data = json.load(check)
```

```python
access_token = data['access_token']
```

```python
### initialization of activities table
url_init = f"https://www.strava.com/api/v3/athlete/activities?access_token={access_token}&per_page=200"
payload={}
headers = {
  'Cookie': '_strava4_session=633nr333v5vqncotecis88kabfigqpff'
}

response = requests.request("GET", url_init, headers=headers, data=payload)
activities_df = pd.json_normalize(response.json())
```

```python
activities_df.iloc[-1]['start_date']
```

```python
s = activities_df.iloc[-1]['start_date']
parsed_date = parser.parse(s)
timestamp = []
timestamp.append(calendar.timegm(parser.parse(s).timetuple()))


#### loop to get all activities
while True:
    url = f"https://www.strava.com/api/v3/athlete/activities?access_token={access_token}&before="+str(timestamp[-1])+"&per_page=200"
    response = requests.request("GET", url, headers=headers, data=payload)
    activities = pd.json_normalize(response.json())
    activities_df = pd.concat([activities_df, activities],ignore_index=True)
    
    s = activities_df.iloc[-1]['start_date']
    parsed_date = parser.parse(s)
    timestamp.append(calendar.timegm(parser.parse(s).timetuple()))
    
    if timestamp[-1]==timestamp[-2]:
        break

```

Weekly

```python
df_weekly = activities_df.groupby(['type','week','year'],as_index=False).agg({'distance': ['sum','max'],'moving_time': ['sum']}).sort_values(by=['year','week'],ascending=[False,False])

df_weekly.columns = ['_'.join(col) for col in df_weekly.columns]
df_weekly.loc[:,['distance_sum','distance_max']]=df_weekly.loc[:,['distance_sum','distance_max']]/1000
df_weekly.loc[:,['moving_time_sum']]=df_weekly.loc[:,['moving_time_sum']]/(60*60)

#activities to keep
some_values=['Run','Ride','Swim']

df_weekly=df_weekly.loc[df_weekly['type_'].isin(some_values)]


df_weekly=df_weekly.pivot_table(df_weekly, ['week_','year_'], 'type_',fill_value=0)#[[cols]]#.sort_index(axis='columns', level='type_')
cols = df_weekly.columns.tolist()

df_weekly

cols = [('distance_sum', 'Ride'), ('distance_sum', 'Run'), ('distance_sum', 'Swim'), ('distance_max', 'Ride'), ('distance_max', 'Run'), ('distance_max', 'Swim'), ('moving_time_sum', 'Ride'), ('moving_time_sum', 'Run'), ('moving_time_sum', 'Swim')]
#df_weekly.loc[:,cols]
```

```python
columns = df_weekly.columns.tolist()
df_weekly.columns = ['_'.join(tuple(reversed(list(col)))) for col in columns]

col_to_drop=[]

df_weekly=df_weekly.reset_index(level=0).reset_index(level=0).sort_values(by=['year_', 'week_'], ascending=True).drop(labels=col_to_drop,axis=1).round(1)
```

```python
df_weekly=df_weekly[['year_', 'week_', 'Ride_distance_sum','Ride_moving_time_sum', 'Ride_distance_max', 
                          'Run_distance_sum','Run_moving_time_sum','Run_distance_max',
                          'Swim_distance_sum', 'Swim_moving_time_sum','Swim_distance_max']
                   ]
```

```python
df_weekly=df_weekly.reset_index(drop=True)
```

```python
import datetime
d = str(df_weekly.year_[0])+'-W'+str(df_weekly.week_[0])
r = datetime.datetime.strptime(d + '-1', "%Y-W%W-%w")
```

```python
df_weekly['date_']=df_weekly.year_.map(str)+'-W'+df_weekly.week_.map(str)+ '-1'
```

```python
df_weekly.insert(2,'date',df_weekly['date_'])
df_weekly['date']=df_weekly['date'].map(lambda date_string: datetime.datetime.strptime(date_string, "%Y-W%W-%w"))
```

```python
import numpy
health[health=='No data']=numpy.nan
health[health=='--']=numpy.nan
```

```python
health.score_sleep=health.score_sleep.astype('float')
health.Duree_sommeil=health.Duree_sommeil.astype('float')
health.FC_repos=health.FC_repos.astype('float')
health.FC_max_day=health.FC_max_day.astype('float')
health.stress=health.stress.astype('float')
health.VO2_max_run=health.VO2_max_run.astype('float')
health.training_load=health.training_load.str.replace(',','').astype('float')

health.loc[:,['Duree_sommeil']]=health.loc[:,['Duree_sommeil']]/(60)
```

```python
health_w=health.groupby(['week','year'],as_index=False).agg({'score_sleep': ['mean'],
                                                    'Duree_sommeil': ['mean'],
                                                    'FC_repos': ['mean'],
                                                    'FC_max_day': ['max'],
                                                    'stress':['mean'],
                                                    'VO2_max_run':['mean'],
                                                    'training_load':['mean']
                                                   
                                                   }).sort_values(by=['year','week'],ascending=[False,False]).round(1)
```

```python
health_w
```

```python
health_w['date_']=health_w.year.map(str)+'-W'+health_w.week.map(str)+ '-1'
```

```python
health_w.insert(2,'date',health_w['date_'])
health_w['date']=health_w['date'].map(lambda date_string: datetime.datetime.strptime(date_string, "%Y-W%W-%w"))
```

```python
columns = health_w.columns.tolist()
health_w.columns = ['_'.join(tuple(reversed(list(col)))) for col in columns]
```
